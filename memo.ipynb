```python
def fujisangezan(location_now=136):
    
    # xの変化量
    x_change = fuji[location_now, 0] - fuji[location_now-1, 0]
    # ｙの変化量
    y_change = fuji[location_now, 3] - fuji[location_now-1,3]
    # 勾配　= ｙの変化量　/ xの変化量
    gradient = y_change / x_change
    
    return gradient, location_now

# ある地点の勾配, 今いる地点
a, b = fujisangezan()
print("今いる地点 = {}, 勾配 = {} ".format(b, a)) 
print(type(int(a)))
print(type(b))
# ハイパーパラメータ

def fujisangezan_x(loc_1=b, α=0.2):
    """ある地点の勾配を計算する関数。
    Parameteres
    -----------
    gradient_now : int
        今いる地点の勾配 
    location_now : int
        今いる地点の番号 
    
    Returns
    -------
    location_move : int
        移動先の地点 
    """
    # 移動先の地点 = 今いる地点 - α×今いる地点の勾配
    location_move = fuji[loc_1, 0] - α * int(a)
    
    return int(location_move)
# 移動先の地点   a:ある地点の勾配
location_move_1 = fujisangezan_x(int(a))


from decimal import Decimal, getcontext, ROUND_HALF_UP
c = getcontext()  # 現在の演算コンテキストを取得
c.rounding = ROUND_HALF_UP  # 丸めモードとして四捨五入を指定
s = location_move_1
d = Decimal(s)

# 移動先の地点(四捨五入)
r = d.quantize(Decimal('0'), rounding=ROUND_HALF_UP)  # 小数点以下を四捨五入
print("移動先の地点 = {} ".format(r)) 
```
