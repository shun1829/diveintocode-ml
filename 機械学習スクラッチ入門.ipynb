```python
import numpy as np
from sklearn.model_selection import train_test_split
import random
```

```python
def scratch_train_test_split(X, y, train_size=0.8):
    """検証データを分割する。
    Parameters
    ----------
    X : ndarray
      訓練データ (n_samples, n_features)
    y : ndarray
      正解値 (n_samples,)
    train_size : float
      何割をtrainとするか指定 (0 < train_size < 1)
    Returns
    -------
    X_train : ndarray
      訓練データ (n_samples, n_features)
    X_test : ndarray
      検証データ (n_samples, n_features)
    y_train : ndarray
      訓練データの正解値 (n_samples,)
    y_test : ndarray
      検証データの正解値 (n_samples,)
    """
    
    arr_rand = np.random.choice(range(X.shape[0]), int(train_size*X.shape[0]))
    split = arr_rand < np.percentile(arr_rand, train_size/10)
    # train_size ×　Xのlen(行) = yのlen(行)をfloatからint型へ変換
    split_point = int(train_size *len(X))
    
    # Xの0行目からsplit_point行目まで抽出
    X_train = X[0:split_point]
    # Xのsplit_point行目から最後の行まで抽出
    X_test = X[split_point:]
    
    # yの0行目からsplit_point行目まで抽出
    y_train= y[0:split_point]
    # yのsplit_point行目から最後の行まで抽出
    y_test= y[split_point:]
    
    pass
    return X_train, X_test, y_train, y_test

x_practice_array = np.arange(100).reshape(50, 2)
y_practice_array = np.arange(50,)
train_size=0.8

# scratch_train_test_split(x_practice_array, y_practice_array, train_size=train_size)
```
